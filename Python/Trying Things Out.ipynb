# Scenario: Library Book Borrowing System
You are writing a program for a small library. The library has a list of books, and each book has a status indicating whether it is available or checked out. The program should:

1. Display all books and their status.
2. Ask the user for their name and age.
3. If the user is under 12, they cannot borrow books (print a message and exit).
4. Otherwise, allow the user to borrow up to 2 books.
5. For each borrowing attempt:
    * Ask the user for the book title.
    * If the book is available, mark it as checked out and print a confirmation.
    * If the book is not available or does not exist, print an appropriate message.
    * After borrowing, display the updated list of books and their status.


````python
# Libraries
import numpy as np

def list_book_availability():
    # List of books and availability
    file = open("List of Books.txt","r")   # "r" has to be in quotes to work
    list_books = file.readlines()
    book_availability = []

    for book in list_books:
        # Randomize book availability
        available = np.random.randint(0,2)
        # Save book and availability in list
        book_availability = book_availability + [[book.strip(), available]]
    return book_availability    # end function, export value

def print_books(book_availability):
    index = 1
    # Print list of books
    for item in book_availability:
        book, available = item
        # Convert availability into string format
        if available == 0:
            available = "Not Available"
        else:
            available = "Available"  
        # Print list of books
        print(f"{index}. {book.strip()}: {available}")
        index += 1
        
def select_book(list_length):
    book_number = input("Enter the number of the book you wish to borrow or press r to leave: ")
    if book_number == "r":
        return book_number
    while not book_number.isdigit() or not (0 < int(book_number) and int(book_number) <= list_length):
        if book_number == "r":
            return book_number
        book_number = input("Value not recognized. Please enter number of book you wish to borrow or press r to leave: ")
    return int(book_number)

def borrow_book(book_number, book_availability):
    # Is book available
    while book_availability[book_number - 1][1] == 0 or type(book_number) == "string": 
        print("Book is not available. Please choose a new book.")
        book_number = select_book(len(book_availability))
        if book_number == "r":
            return book_number, book_availability
    # Checkout book
    print(f"{book_availability[book_number - 1][0]} has been checked out.")
    book_availability[book_number - 1][1] = 0
    return book_number, book_availability
    
# List books and availability
book_availability = list_book_availability()

# Ask user for name and age.
name = input("Enter your name: ")
age = input("Enter your age: ")
while not age.isdigit() or not (0 < int(age) and int(age) < 99):
    age = input("Age not recognized. Enter your age: ")
checkout = 2
book_number = ""

# Verify user is age 12+
if int(age) >= 12:
    print(f"\nWelcome {name}.")
    while checkout > 0 and book_number != "r":
        print(f"You may borrow up to {checkout} book(s).\n")
        print_books(book_availability)
        book_number = select_book(len(book_availability))
        if book_number == "r":
            break
        book_number, book_availability = borrow_book(book_number, book_availability)
        checkout -= 1
        print()
    print(f"Now leaving. Goodbye {name}.")
else:
    print(f"You are too young to check out books.\nNow leaving. Goodbye {name}.")
````

## camelCase
Given a string, return its camel case version using the following rules:
- Words in the string argument are separated by one or more characters from the following set: space ( ), dash (-), or underscore (_). Treat any sequence of these as a word break.
- The first word should be all lowercase.
- Each subsequent word should start with an uppercase letter, with the rest of it lowercase.
- All spaces and separators should be removed.

### [Split a string on multiple delimiters in Python](https://www.geeksforgeeks.org/python/python-split-multiple-characters-from-string/)

- The `re.split()` function from the `re` module is the most straightforward way to split a string on multiple delimiters. It uses a regular expression to define the delimiters.
    - `[;,\s]+`: This pattern matches one or more occurrences of a semicolon (;), comma (,), or whitespace (\s).
    - `re.split`: Splits the string wherever the pattern matches.
```
import re

s = "apple, banana; orange grape"

# Split using re.split
res = re.split(r'[;,\s]+', s)
print(res)
```
- Using `translate()` and `split()`
- If the delimiters are fixed and limited to a set of characters then we can replace them with a single delimiter (like a space) using `translate()` and use `split()`.
    - `str.maketrans`: Creates a translation map where , and ; are replaced by a space.
    - `str.translate`: Applies the translation map to the string.
    - `str.split`: Splits the string on whitespace
```
s = "apple, banana; orange grape"

# Replace delimiters with a space and split
s = s.translate(str.maketrans({',': ' ', ';': ' '}))
res = s.split()
print(res)
```
- Chaining `replace()` and `split()`

````python
def to_camel_case(s):
    # string is input. use .lower(), .upper(), re.split()
    import re
    string = re.split(r'[-_\s]+',s)

    return string

to_camel_case("test_phrase-with delimiters")
````
